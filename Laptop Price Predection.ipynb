```python
# Define the categories
categories = {
    'resolution': ['Full HD', '4K Ultra HD', 'Quad HD+','Retina Display'],
    'type': ['IPS Panel', 'Touchscreen']
}

# Create new columns for each category
for category, values in categories.items():
    for value in values:
        df[f'{category}_{value.replace(" ", "_")}'] = df['ScreenResolution'].apply(lambda x: 1 if value in x else 0)

# Extract resolution
df['resolution_width'] = df['ScreenResolution'].apply(lambda x: x.split()[-1].split('x')[0])
df['resolution_height'] = df['ScreenResolution'].apply(lambda x: x.split()[-1].split('x')[1])

df
```

```python
df=df.drop(columns="ScreenResolution")
```

```python
df['Cpu'].unique()
```

```python
# Extract company name, model name, and clock speed
df['Company_Name'] = df['Cpu'].apply(lambda x: x.split()[0])
df['Model_Name'] = df['Cpu'].apply(lambda x: ' '.join(x.split()[1:-1]))
df['Clock_Speed'] = df['Cpu'].apply(lambda x: float(x.split()[-1].replace('GHz', '')))

```

```python
df['Model_Name'].unique()
```

```python
import pandas as pd
import re

# Define a function to remove series number
def remove_series_number(model_name):
    return re.sub(r'\s*\d+.*', '', model_name)

# Apply the function to the 'CPU' column
df['Model_Name'] = df['Model_Name'].apply(remove_series_number)

```

```python
df['Weight'].unique()
```

```python
# Define the categories
categories = {
    'brand': ['Intel', 'AMD', 'Samsung'],
    'type': ['Core i3', 'Core i5', 'Core i7', 'A9-Series', 'E-Series', 'A6-Series', 'Celeron Dual Core', 'Ryzen', 'Pentium Dual Core', 'Pentium Quad Core', 'Atom', 'Xeon', 'FX', 'A4-Series', 'A10-Series', 'A12-Series']
}

# Create new columns for each category
for category, values in categories.items():
    for value in values:
        df[f'{category}_{value.replace(" ", "_")}'] = df['Cpu'].apply(lambda x: 1 if value in x else 0)

# Extract clock speed
df['clock_speed_GHz'] = df['Cpu'].apply(lambda x: float(x.split()[-1].replace('GHz', '')))
```
